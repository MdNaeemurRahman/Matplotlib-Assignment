## 1. Import Matplotlib & Numpy

```python
import matplotlib.pyplot as plt
import numpy as np
```

## 2. Preparing data

Create x vector which has 100 elements and the element values gradually increase from 0 to 99.

Randomly create y vector which has 100 elements and the element values are randomly generated in the range of [0, 1000)

```python
x = np.arange(0, 100)
y = np.random.randint(low = 0 , high = 1000 , size = 100)
```

## 3. Draw a line chart

Use the above data to draw a line chart.

```python
plt.plot(x,y)

plt.show()
```

## 4. Config the axes

Use the above generated dataset, let set the name for x-axis, y-axis and the legend of your chart.

```python
fig = plt.figure(figsize = (10,6))
plt.plot(x,y , label = "Random Number", marker = ".")
plt.xlabel("X-Axis", size = 18)
plt.ylabel("Y-Axis" , size = 18)
plt.title("Line Chart" , size = 23, fontdict={"fontname" : "Arial"})
plt.legend()
plt.show()
```

## 5. Multiple line chart

Create x vector which has 100 elements and the element values gradually increase from 0 to 99.

Randomly create y1 and y2 vector which have 100 elements and the element values are randomly generated in the range of [0, 1000)

Let draw the above two lines y1 and y2. The chart must has:

names of x-axis and y-axis
chart legend
different colors for the lines of y1 and y2

```python
x = np.arange(0,100)
y1 = np.random.randint(low = 0 , high = 1000, size = 100)
y2 = np.random.randint(low = 0 , high = 1000, size = 100)

fig = plt.figure(figsize=(14, 6))
plt.plot(x, y1, "-b", label="y1", marker = "o")
plt.plot(x, y2, ":r", label="y2", marker = "^")
plt.legend(loc="lower right")
plt.title("Multiple Line Chart", size = 20 , fontdict={"fontname" : "Arial"})
plt.xlabel("X-Axis", size = 18)
plt.ylabel("Y-Axis", size = 18)

plt.show()
```

## 6. Draw a box plot

Use the above y1, y2 vectors to draw a box plot.


```python
data_to_plot = [y1,y2]

fig = plt.figure(figsize=(12, 6))
labels = ["Y1", "Y2" ]

plt.ylabel("Observed values" , size = 12)
plt.grid(True)
plt.title("Box Plot" , size = 20)
plt.boxplot(data_to_plot, labels = labels)

plt.show()
```

## 7. Draw a histogram

```python
w = np.random.standard_normal(1000) 
z = np.random.normal(1,1,1000)

fig = plt.figure(figsize=(10, 5))
plt.hist(w , bins= 25, color= "b" , edgecolor = "g", alpha=.8)
plt.hist(z , bins= 25, color= "g" , edgecolor = "g", alpha=.8)
plt.title("Histogram" , size = 18)
plt.xlabel("Size of Number", size = 13)
plt.ylabel("Number of Observation", size = 13)
plt.grid(axis="y", alpha=.6)
plt.show()

fig = plt.figure(figsize=(10, 5))
plt.title("Cumulative Histogram" , size = 18)
plt.hist(w , cumulative= True)
plt.grid(axis="y")
plt.show()
```

## 8. Draw a bar plot

```python
company=['GOOGLE','AMAZON','APPLE','FB']
revenue=[90,136,89,27]
profit=[40,20,34,12] 
```

```python
fig = plt.figure(figsize=(12, 6))

xpos = np.arange(len(company))

plt.bar(xpos - 0.20, revenue, width=-0.4, label="Revenue")
plt.bar(xpos + 0.20, profit, width=0.4, label="Profit", color = "red")

# plt.bar(xpos, revenue)
# plt.bar(xpos, profit)

plt.xticks(xpos, company , size = 12)
plt.ylabel("Profit and Revenue", size = 13)
plt.title("Tech Company Profit and Revenue" , size = 20)
plt.legend()
plt.show()
```

## 9. Draw a scatter plot

Use the above x, y1, y2 vectors to draw a scatter plot

```python
fig = plt.figure(figsize=(12, 5))

plt.scatter(x, y1 , label = "Y1", marker = "^")
plt.scatter(x, y2, label = "Y2")
plt.xlabel("X-Axis" , size = 17)
plt.ylabel("Y-Axis", size = 17)
plt.title("Scatter Plot" , size = 20)
plt.legend(fontsize= 13 , loc = "upper right")
plt.show()
```
